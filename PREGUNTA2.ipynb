```python
import csv

# Leer archivo CSV
def leer_csv(filepath):
    with open(filepath, newline='') as csvfile:
        reader = csv.reader(csvfile)
        datos = [row for row in reader]  # Lee cada fila como una lista
    return datos

# Extraer todas las columnas de los datos
def extraer_columnas(datos):
    columnas = []
    for i in range(len(datos[0])):  # Iterar sobre cada índice de columna
        columna = [float(row[i]) for row in datos[1:]]  # Convertir a float, omitiendo el encabezado
        columnas.append(columna)
    return columnas

# Función para calcular percentil
def calcular_percentil(datos, p):
    datos_ordenados = sorted(datos)
    n = len(datos)

    # Posición del percentil
    P = (p / 100) * (n - 1) + 1
    P = int(P)  # Redondeamos al entero más cercano
    return datos_ordenados[P-1]

# Leer y procesar el CSV
filepath = '/content/Drive/MyDrive/datos/diabetes.csv'  # Asegúrate de usar la ruta correcta
datos = leer_csv(filepath)

# Extraer todas las columnas
columnas = extraer_columnas(datos)

# Calcular percentiles y cuartiles para cada columna
for i, columna in enumerate(columnas):
    percentil_100 = calcular_percentil(columna, 100)
    cuartil_1 = calcular_percentil(columna, 25)
    cuartil_2 = calcular_percentil(columna, 50)
    cuartil_3 = calcular_percentil(columna, 75)

    print(f"Columna {i+1}:")
    print(f"  Percentil 100: {percentil_100}")
    print(f"  Cuartil 1: {cuartil_1}")
    print(f"  Mediana (Percentil 50): {cuartil_2}")
    print(f"  Cuartil 3: {cuartil_3}")
    print()
```

DISTRIBUCION DE BERNOULLI

Este tipo de análisis es útil cuando quieres evaluar eventos binarios, como en este caso, si una persona tiene o no diabetes.

```python
#distribucion de bernoulli
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Drive/MyDrive/datos/diabetes.csv')
outcome = df['Outcome']

# Ver los datos únicos y su frecuencia
print(outcome.value_counts())

# Graficar la distribución de Bernoulli
plt.hist(outcome, bins=2, density=True, alpha=0.6, color='b')
plt.title("Distribución Bernoulli (Diabetes Outcome)")
plt.xlabel("Outcome (0 = No Diabetes, 1 = Diabetes)")
plt.ylabel("Frecuencia")
plt.xticks([0, 1])  # Mostrar solo las etiquetas 0 y 1
plt.show()
```

Las columnas que seleccionaremos son:

**Glucose:** Este atributo representa el nivel de glucosa en sangre de los pacientes.

**IMC (Body Mass Index):** Este atributo mide la relación entre el peso y la altura de una persona, proporcionando información sobre el estado nutricional.

**Age:** Este atributo indica la edad de los pacientes, lo cual es relevante para entender la relación entre la diabetes y la edad.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el CSV
df = pd.read_csv('/content/Drive/MyDrive/datos/diabetes.csv')

# Seleccionar columnas relevantes
selected_columns = df[['Glucose', 'BMI', 'Age']]

# Calcular características para cada columna
for column in selected_columns.columns:
    mean_value = selected_columns[column].mean()
    std_dev = selected_columns[column].std()
    min_value = selected_columns[column].min()
    max_value = selected_columns[column].max()

    print(f"--- {column} ---")
    print(f"Media: {mean_value}")
    print(f"Desviación Estándar: {std_dev}")
    print(f"Valor Mínimo: {min_value}")
    print(f"Valor Máximo: {max_value}")
    print("\n")

# Graficar un mapa de calor de correlación
plt.figure(figsize=(10, 6))
correlation = selected_columns.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Mapa de Calor de Correlación entre Glucose, IMC y Age")
plt.show()

# Graficar dispersión
plt.figure(figsize=(15, 5))

# Dispersión Glucose vs. BMI
plt.subplot(1, 2, 1)
plt.scatter(df['Glucose'], df['BMI'], alpha=0.5, color='blue')
plt.title("Dispersión entre Glucose y IMC")
plt.xlabel("Glucose")
plt.ylabel("IMC")

# Dispersión Glucose vs. Age
plt.subplot(1, 2, 2)
plt.scatter(df['Glucose'], df['Age'], alpha=0.5, color='green')
plt.title("Dispersión entre Glucose y Age")
plt.xlabel("Glucose")
plt.ylabel("Age")

plt.tight_layout()
plt.show()
```
